# Plot DEG by subclass

```python
import pandas as pd
import os
from pathlib import Path
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import matplotlib
```

```python
import scanpy as sc
import decoupler as dc
```

```python
os.chdir("/tscc/projects/ps-epigen/users/biy022/biccn/data/SNAREdata/")
source_dir = Path("scenicplus/other_celltypes/")
deg_dir = Path("regional_dar_deg/deg")
```

```python
subclass = "L4_IT"
rna_counts = sc.read_h5ad(source_dir / "{}_rna_matrix.h5ad".format(subclass))
if rna_counts.raw != None:
    rna_counts = rna_counts.raw.to_adata().copy()
rna_counts_aggr = dc.get_pseudobulk(
    rna_counts,
    sample_col="Region",
    groups_col=None,
    min_cells=0,
    min_counts=0,
    remove_empty=False
)
sc.pp.normalize_total(rna_counts_aggr, target_sum=1e6)
sc.pp.log1p(rna_counts_aggr, base=2)
```

```python
regional_degs = dict()
region_order = ["A24", "A9", "FI", "M1C", "S1C", "MTG", "A1C", "AnG", "V1C"]
for region in region_order:
    deg_table = pd.read_csv(deg_dir / subclass / "{}_result.tsv".format(region), sep="\t", header=0, index_col=0)
    deg_sig_list = deg_table[(deg_table["logFC"] > 0) & (deg_table["FDR"] < 0.001)].index.to_list()
    regional_degs[region] = deg_sig_list
```

```python
deg_list = [xx for region in region_order for xx in regional_degs[region]]
```

```python
figure_dir = Path("/tscc/projects/ps-epigen/users/biy022/biccn/analysis/final_figures/")
```

```python
plot_mat = rna_counts_aggr[region_order, deg_list].X.T
plot_mat_mean = np.mean(plot_mat, axis=1, keepdims=True)
plot_mat_std = np.std(plot_mat, axis=1, keepdims=True)
plot_mat_z = (plot_mat - plot_mat_mean) / plot_mat_std
fig, ax = plt.subplots(1, 1)
fig.set_figwidth(4)
fig.set_figheight(5)

plt.rcParams["font.family"] = "Ubuntu Mono"
sns.heatmap(
    plot_mat_z,
    cmap="coolwarm",
    vmin=-2,
    vmax=2,
    ax=ax,
    yticklabels=[],
    xticklabels=region_order
)
ax.set_title(subclass, fontsize=20)
plt.tight_layout()
plt.show()
fig.savefig(figure_dir / "{}-deg-heatmap.pdf".format(subclass), dpi=300, format="pdf")
```
