## RNA (correlation with order along the axis)

```python
L5_ET_rna = sc.read_h5ad("L5_ET_rna_matrix.h5ad")
L5_ET_rna_counts = L5_ET_rna.raw.to_adata()
```

```python
L5_ET_rna_region_counts = dc.get_pseudobulk(
    adata=L5_ET_rna_counts,
    sample_col="Region",
    groups_col=None,
    min_cells=0,
    min_counts=0,
    min_prop=0.0,
    min_smpls=0,
    remove_empty=True,
    mode="sum"
)
```

```python
sc.pp.normalize_total(L5_ET_rna_region_counts, target_sum=1e6)
sc.pp.log1p(L5_ET_rna_region_counts, base=2)
```

```python
region_align = ["A24", "A9", "FI", "M1C", "S1C", "MTG", "A1C", "AnG", "V1C"]
data_matrix = L5_ET_rna_region_counts[region_align, :].X
```

```python
import scipy.stats as stats
```

```python
b = np.arange(0, len(region_align))
corrs = np.zeros(data_matrix.shape[1])
pvalues = np.zeros(data_matrix.shape[1])

for i in range(0, data_matrix.shape[1]):
    a = data_matrix[:, i]
    corr, pvalue = stats.spearmanr(a, b)
    corrs[i] = corr
    pvalues[i] = pvalue
```

```python
padjusted = stats.false_discovery_control(pvalues, method="bh")
```

```python
spearman_result = pd.DataFrame({
    "gene": L5_ET_rna_region_counts.var.index.to_list(),
    "corr": corrs,
    "p_value": pvalues,
    "p_adjusted": padjusted,
})
```

```python
spearman_result.to_csv(
    "L5_ET_rostral_caudal_spearman_corr.tsv", sep="\t", header=True, index=False)
```

```python
spearman_sig_positive = spearman_result[
    (spearman_result["p_adjusted"] < 0.05) & (spearman_result["corr"] > 0)]
spearman_sig_negative = spearman_result[
    (spearman_result["p_adjusted"] < 0.05) & (spearman_result["corr"] < 0)]
```

```python
positive_sig_genes = spearman_sig_positive["gene"].to_list()
positive_matrix = L5_ET_rna_region_counts[region_align, positive_sig_genes].X
```

```python
sns.clustermap(
    positive_matrix.T,
    xticklabels=region_align,
    yticklabels=[],
    z_score="row",
    figsize=(8, 8),
    col_cluster=False,
    dendrogram_ratio=(0.1, 0.05),
    cbar_pos=(0.5, 0.975, 0.2, 0.02),
    cbar_kws={"orientation": "horizontal"},
    tree_kws={"linewidth": 2},
    # linewidth=2,
    # linecolor="black",
)
```

```python
negative_sig_genes = spearman_sig_negative["gene"].to_list()
negative_matrix = L5_ET_rna_region_counts[region_align, negative_sig_genes].X
```

```python
sns.clustermap(
    negative_matrix.T,
    xticklabels=region_align,
    yticklabels=[],
    z_score="row",
    figsize=(8, 10),
    col_cluster=False,
    dendrogram_ratio=(0.1, 0.05),
    cbar_pos=(0.5, 0.975, 0.2, 0.02),
    cbar_kws={"orientation": "horizontal"},
    tree_kws={"linewidth": 2},
    linewidth=0.5,
    linecolor="black",
)
```

```python
full_gene = spearman_result[spearman_result["p_adjusted"] < 0.05].sort_values(
    by="corr", ascending=False)["gene"]
full_sorted_matrix = L5_ET_rna_region_counts[region_align, full_gene].X
sns.clustermap(
    full_sorted_matrix.T,
    xticklabels=region_align,
    yticklabels=[],
    z_score="row",
    figsize=(8, 6),
    col_cluster=False,
    row_cluster=False,
    dendrogram_ratio=(0.1, 0.05),
    cbar_pos=(0.5, 0.975, 0.2, 0.02),
    cbar_kws={"orientation": "horizontal"},
    # tree_kws={"linewidth": 2},
    # linewidth=0.5,
    # linecolor="black",
)
```

```python
positive_sig_genes = spearman_result[
    (spearman_result["corr"] > 0) & (spearman_result["p_adjusted"] < 0.05)]["gene"].to_list()
with open("rostral_caudal_spearman/positive_genes_bh_5e-2.tsv", "w") as fout:
    for gene in positive_sig_genes:
        fout.write("{}\n".format(gene))
```

```python
negative_sig_genes = spearman_result[
    (spearman_result["corr"] < 0) & (spearman_result["p_adjusted"] < 0.05)]["gene"].to_list()
with open("rostral_caudal_spearman/negative_genes_bh_5e-2.tsv", "w") as fout:
    for gene in negative_sig_genes:
        fout.write("{}\n".format(gene))
```

```python
full_gene = spearman_result["gene"].to_list()
with open("rostral_caudal_spearman/full_gene_list.tsv", "w") as fout:
    for gene in full_gene:
        fout.write("{}\n".format(gene))
```

## ATAC spearman with index

```python
L5_ET_atac = sc.read_h5ad("L5_ET_atac_matrix.h5ad")
```

```python
L5_ET_atac_region_counts = dc.get_pseudobulk(
    adata=L5_ET_atac,
    sample_col="Region",
    groups_col=None,
    min_cells=0,
    min_counts=0,
    min_prop=0.0,
    min_smpls=0,
    remove_empty=True,
    mode="sum"
)
```

```python
sc.pp.normalize_total(L5_ET_atac_region_counts, target_sum=1e6)
sc.pp.log1p(L5_ET_atac_region_counts, base=2)
```

```python
region_align = ["A24", "A9", "FI", "M1C", "S1C", "MTG", "A1C", "AnG", "V1C"]
L5_ET_atac_region_counts.obs["Region"] = pd.Categorical(
    L5_ET_atac_region_counts.obs["Region"], categories=region_align
)
```

```python
b = L5_ET_atac_region_counts.obs.loc[region_align, "Region"].cat.codes.to_list()
data_matrix = L5_ET_atac_region_counts[region_align, :].X
corrs = np.zeros(data_matrix.shape[1])
pvalues = np.zeros(data_matrix.shape[1])

for i in range(0, data_matrix.shape[1]):
    a = data_matrix[:, i]
    corr, pvalue = stats.spearmanr(a, b)
    corrs[i] = corr
    pvalues[i] = pvalue
```

```python
padjusted = stats.false_discovery_control(pvalues, method="bh")
```

```python
spearman_result = pd.DataFrame({
    "region": L5_ET_atac_region_counts.var.index.to_list(),
    "corr": corrs,
    "p_value": pvalues,
    "p_adjusted": padjusted,
})
```

```python
spearman_result.to_csv(
    "L5_ET_atac_rostral_caudal_spearman_corr.tsv", sep="\t", header=True, index=False)
```

```python
spearman_sig_positive = spearman_result[
    (spearman_result["p_adjusted"] < 0.1) & (spearman_result["corr"] > 0)]
spearman_sig_negative = spearman_result[
    (spearman_result["p_adjusted"] < 0.1) & (spearman_result["corr"] < 0)]
```
