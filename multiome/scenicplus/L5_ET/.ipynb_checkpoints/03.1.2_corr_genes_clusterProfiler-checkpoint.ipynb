# Use clusterProfiler GO term analysis

```r
suppressPackageStartupMessages({
    library(clusterProfiler)
    library(ggplot2)
    library(fs)
    library(tidyverse)
    library(org.Hs.eg.db)
})
```

```r
setwd("/tscc/projects/ps-epigen/users/biy022/biccn/data/SNAREdata/scenicplus/L5_ET/")
```

```r
background.file <- "rostral_caudal_spearman/full_gene_list.tsv"
background.table <- read.table(background.file, header = FALSE, sep = "\t")
background.list <- background.table$V1
```

```r
### Positive correlated genes
```

```r
curr.table <- read.table(
    "rostral_caudal_spearman/positive_genes_bh_7e-1_5e-2_donor_replicate.tsv",
    header = FALSE, sep = "\t")
curr.list <- curr.table$V1

curr.result <- enrichGO(
    gene = curr.list,
    universe = background.list,
    keyType = "SYMBOL",
    OrgDb = org.Hs.eg.db,
    ont = "BP",
    pvalueCutoff = 0.1,
    readable = TRUE,
)
curr.result.simplified <- clusterProfiler::simplify(curr.result)
output.file <- "rostral_caudal_spearman/positive_genes_bh_7e-1_5e-2_donor_replicate.gsea.tsv"
write.table(
    curr.result.simplified@result,
    output.file,
    quote = FALSE,
    col.names = TRUE,
    row.names = TRUE,
    sep = "\t"
)
```

```r
curr.table <- read.table(
    "rostral_caudal_spearman/negative_genes_bh_7e-1_5e-2_donor_replicate.tsv",
    header = FALSE, sep = "\t")
curr.list <- curr.table$V1

curr.result <- enrichGO(
    gene = curr.list,
    universe = background.list,
    keyType = "SYMBOL",
    OrgDb = org.Hs.eg.db,
    ont = "BP",
    pvalueCutoff = 0.1,
    readable = TRUE,
)
curr.result.simplified <- clusterProfiler::simplify(curr.result)
output.file <- "rostral_caudal_spearman/negative_genes_bh_7e-1_5e-2_donor_replicate.gsea.tsv"
write.table(
    curr.result.simplified@result,
    output.file,
    quote = FALSE,
    col.names = TRUE,
    row.names = TRUE,
    sep = "\t"
)
```

## Plot the result

```r
result_table <- read.table(
    "rostral_caudal_spearman/negative_genes_bh_7e-1_5e-2_donor_replicate.gsea.tsv",
    header = TRUE, sep = "\t", row.names = 1
)
```

```r
plot_dir <- "../../../../analysis/scenicplus/L5_ET/figures/"
```

```r
options(repr.plot.width = 16, repr.plot.height = 12)
result_table$nlogpvalue <- -log10(result_table$pvalue)
result_table$ID <- factor(result_table$ID, levels = rev(result_table$ID))
plot <- ggplot(result_table, aes(x=ID, y=nlogpvalue)) +
    geom_bar(stat = "identity", fill = "skyblue", colour = "black", linewidth = 1) +
    geom_text(aes(label = Description), hjust = 1.05, size=5) +
    coord_flip() +
    theme_classic() +
    theme(
        axis.title.x = element_text(size = 20),
        axis.title.y = element_text(size = 20),
        axis.text.x = element_text(size=17.5),
        axis.text.y = element_blank()
    )
ggsave(
    plot = plot,
    filename = "L5_ET_corr_donor_replicate_negative_RNA_GSEA.png",
    path = plot_dir,
    width = 16,
    height = 12,
    dpi = 300,
    device = "png"
)
plot
```
