# Use clusterProfiler GO term analysis

```r
suppressPackageStartupMessages({
    library(clusterProfiler)
    library(ggplot2)
    library(fs)
    library(tidyverse)
    library(org.Hs.eg.db)
})
```

```r
setwd("/tscc/projects/ps-epigen/users/biy022/biccn/data/SNAREdata/scenicplus/L4_IT/")
```

```r
background.file <- "rostral_caudal_spearman_expr_filtered//full_gene_list.tsv"
background.table <- read.table(background.file, header = FALSE, sep = "\t")
background.list <- background.table$V1
```

```r
### Positive correlated genes
```

```r
curr.table <- read.table(
    "rostral_caudal_spearman_expr_filtered/positive_corr_7e-1_bh_1e-2_donor.tsv",
    header = FALSE, sep = "\t")
curr.list <- curr.table$V1

curr.result <- enrichGO(
    gene = curr.list,
    universe = background.list,
    keyType = "SYMBOL",
    OrgDb = org.Hs.eg.db,
    ont = "BP",
    pvalueCutoff = 0.1,
    readable = TRUE,
)
curr.result.simplified <- clusterProfiler::simplify(curr.result)
output.file <- "rostral_caudal_spearman_expr_filtered/positive_corr_7e-1_bh_1e-2_donor_gsea.tsv"
write.table(
    curr.result.simplified@result,
    output.file,
    quote = FALSE,
    col.names = TRUE,
    row.names = TRUE,
    sep = "\t"
)
```

```r
curr.table <- read.table(
    "rostral_caudal_spearman_expr_filtered/negative_corr_7e-1_bh_1e-2_donor.tsv",
    header = FALSE, sep = "\t")
curr.list <- curr.table$V1

curr.result <- enrichGO(
    gene = curr.list,
    universe = background.list,
    keyType = "SYMBOL",
    OrgDb = org.Hs.eg.db,
    ont = "BP",
    pvalueCutoff = 0.1,
    readable = TRUE,
)
curr.result.simplified <- clusterProfiler::simplify(curr.result)
output.file <- "rostral_caudal_spearman_expr_filtered/negative_corr_7e-1_bh_1e-2_donor_gsea.tsv"
write.table(
    curr.result.simplified@result,
    output.file,
    quote = FALSE,
    col.names = TRUE,
    row.names = TRUE,
    sep = "\t"
)
```

## Plot the result

```r
result_table <- read.table(
    "rostral_caudal_spearman_expr_filtered/positive_corr_7e-1_bh_1e-2_donor_gsea.tsv",
    header = TRUE, sep = "\t", row.names = 1
)
```

```r
plot_dir <- "../../../../analysis/scenicplus/L4_IT/figures/expr_filtered/"
```

```r
options(repr.plot.width = 16, repr.plot.height = 3)
result_table$nlogpvalue <- -log10(result_table$pvalue)
result_table$ID <- factor(result_table$ID, levels = rev(result_table$ID))
plot <- ggplot(result_table, aes(x=ID, y=nlogpvalue)) +
    geom_bar(stat = "identity", fill = "skyblue", colour = "black", linewidth = 1) +
    geom_text(aes(label = Description), hjust = 1.05, size=5) +
    coord_flip() +
    theme_classic() +
    theme(
        axis.title.x = element_text(size = 20),
        axis.title.y = element_text(size = 20),
        axis.text.x = element_text(size=17.5),
        axis.text.y = element_blank()
    )
ggsave(
    plot = plot,
    filename = "L4_IT_positive_RNA_GSEA.png",
    path = plot_dir,
    width = 16,
    height = 3,
    dpi = 300,
    device = "png"
)
plot
```

# GSEA all celltypes (padjusted < 0.05, |corr| > 0.7)

```r
suppressPackageStartupMessages({
    library(clusterProfiler)
    library(ggplot2)
    library(fs)
    library(tidyverse)
    library(org.Hs.eg.db)
})
```

```r
setwd("/tscc/projects/ps-epigen/users/biy022/biccn/data/SNAREdata/")
setwd("scenicplus/other_celltypes/corr_genes_7e-1_5e-2//")
```

```r
for (file in dir_ls("./", glob = "*full*")) {
    celltype <- str_split_1(file, "_full_list.tsv")[1]
    print(celltype)
    flush.console()
    
    background_file <- file
    background_table <- read.table(background_file, header = FALSE, sep = "\t")
    background_list <- background_table$V1
    
    if (file_size(sprintf("%s_positive.tsv", celltype)) <= 0) {
        next
    }
    curr_table <- read.table(sprintf("%s_positive.tsv", celltype), header = FALSE, sep = "\t")
    curr_list <- curr_table$V1

    curr_result <- enrichGO(
        gene = curr_list,
        universe = background_list,
        keyType = "SYMBOL",
        OrgDb = org.Hs.eg.db,
        ont = "BP",
        pvalueCutoff = 0.1,
        readable = TRUE,
    )
    if (is.null(curr_result)) {
        next
    }
    
    curr_result_simplified <- clusterProfiler::simplify(curr_result)
    output_file <- sprintf("%s_positive_gsea.tsv", celltype)
    write.table(
        curr_result_simplified@result,
        output_file,
        quote = FALSE,
        col.names = TRUE,
        row.names = TRUE,
        sep = "\t"
    )
}
```
