# Select networks for selected genes

```python
import os
from pathlib import Path
root_dir = Path("/tscc/projects/ps-epigen/users/biy022/biccn/data/SNAREdata/scenicplus/")
os.chdir(root_dir / "L4_IT")
```

```python
import pandas as pd
import numpy as np
import math
from matplotlib_venn import venn2, venn2_circles
from matplotlib import rcParams
import matplotlib.pyplot as plt
```

```python
direct_table = pd.read_csv("scplus_outputs/eRegulon_direct.tsv", sep="\t", header=0)
extend_table = pd.read_csv("scplus_outputs/eRegulons_extended.tsv", sep="\t", header=0)
```

```python
regulons_table = pd.concat([direct_table, extend_table], axis=0)
regulons_table.reset_index(drop=True, inplace=True)
```

```python
RC_positive_peak_table = pd.read_csv(
    "../other_celltypes/corr_regions_5e-1_5e-2_expr_filtered//L4_IT_positive.tsv",
    sep="\t", header=None
)
RC_positive_peak_table.columns = ["Region"]
RC_negative_peak_table = pd.read_csv(
    "../other_celltypes/corr_regions_5e-1_5e-2_expr_filtered//L4_IT_negative.tsv",
    sep="\t", header=None
)
RC_negative_peak_table.columns = ["Region"]
```

```python
RC_positive_peak_table["Region"] = RC_positive_peak_table.apply(lambda x: ":".join(x["Region"].split("-", 1)), axis=1)
RC_negative_peak_table["Region"] = RC_negative_peak_table.apply(lambda x: ":".join(x["Region"].split("-", 1)), axis=1)
```

## Calcium related genes

```python
calcium_gene_table = pd.read_csv("../../share/genes_of_interest/calcium_genes.tsv", sep="\t", header=None)
calcium_gene_table.columns = ["Gene"]
```

```python
all_edges = pd.merge(regulons_table, calcium_gene_table, on=["Gene"], how="inner")
positive_edges = pd.merge(all_edges, RC_positive_peak_table, on="Region", how="inner")
negative_edges = pd.merge(all_edges, RC_negative_peak_table, on="Region", how="inner")
```

```python
all_edges.to_csv("network_of_interest/L4_IT_network_calcium_genes.tsv", sep="\t", header=True, index=False)
positive_edges.to_csv("network_of_interest/L4_IT_network_RC_positive_calcium_genes.tsv", sep="\t", header=True, index=False)
negative_edges.to_csv("network_of_interest/L4_IT_network_RC_negative_calcium_genes.tsv", sep="\t", header=True, index=False)
```

```python
RC_atac_corr_scores["region"] = RC_atac_corr_scores.apply(lambda x: ":".join(x["region"].split("-", 1)), axis=1)
```

```python
RC_atac_corr_scores["group"] = "NULL_DIST"
RC_atac_corr_scores.loc[RC_atac_corr_scores["region"].isin(all_edges["Region"]), "group"] = "SELECTED"
```

```python
motif_database = pd.read_csv(
    "/tscc/projects/ps-epigen/users/biy022/MGH/data/metadata/homer2_motif_database.csv",
    header=0, sep=","
)
motif_database = motif_database.dropna()
homer_result_positive = pd.read_csv(
    "/tscc/projects/ps-epigen/users/biy022/biccn/data/SNAREdata/scenicplus/L4_IT/rostral_caudal_spearman_expr_filtered/homer/positive_correlation/knownResults.txt",
    header=0, sep="\t"
)
homer_result_positive = homer_result_positive[homer_result_positive["q-value (Benjamini)"] < 0.01]
homer_positive_tfs = motif_database[motif_database["motif_names"].isin(homer_result_positive["Motif Name"])]["tfs"].to_list()
homer_positive_tfs.extend([
    "ETS1", "ELF2", "JDP2", "TCF12", "PLAGL1", "KLF12", "ETS2", "ETV5",
    "KLF6", "RORA", "ARNTL", "NFIB", "NR2F1", "LHX2", "BACH2", "NFIX"
])
homer_positive_tfs = set(homer_positive_tfs)
```

```python
pd.options.mode.chained_assignment = None 
import scanpy as sc
import decoupler as dc
import numpy as np
import scipy.stats as stats
import matplotlib.patches as patches
```

```python
## Plot the expression gradient of the intersected TFs
L4_IT_rna = sc.read_h5ad("L4_IT_rna_matrix.h5ad")
L4_IT_rna_counts = L4_IT_rna.raw.to_adata()
L4_IT_rna_region_counts = dc.get_pseudobulk(
    adata=L4_IT_rna_counts,
    sample_col="Region",
    groups_col=None,
    min_cells=0,
    min_counts=0,
    min_prop=0.0,
    min_smpls=0,
    remove_empty=True,
    mode="sum"
)
sc.pp.normalize_total(L4_IT_rna_region_counts, target_sum=1e6)
sc.pp.log1p(L4_IT_rna_region_counts, base=2)
region_align = ["A24", "A9", "FI", "M1C", "S1C", "MTG", "A1C", "AnG", "V1C"]

relevant_tfs = list(geneset_tfs)
plot_matrix = L4_IT_rna_region_counts[region_align, relevant_tfs].X
g = sns.clustermap(
    plot_matrix.T,
    xticklabels=region_align,
    yticklabels=relevant_tfs,
    z_score="row",
    figsize=(8, 15),
    col_cluster=False,
    dendrogram_ratio=(0.1, 0.05),
    cbar_pos=(0.5, 0.975, 0.2, 0.02),
    cbar_kws={"orientation": "horizontal"},
    tree_kws={"linewidth": 2},
    linewidth=2,
    linecolor="black",
)
ax = g.ax_heatmap
ax.tick_params(axis="x", labelsize=15)
ax.tick_params(axis="y", labelsize=15)
pos = ax.get_position()
x0 = pos.x0
y0 = pos.y0
width = pos.width
height = pos.height
rect = patches.Rectangle(
    (x0, y0),
    width,
    height,
    linewidth=2, 
    edgecolor="black", 
    facecolor="none", 
    transform=plt.gcf().transFigure
)
plt.gcf().add_artist(rect)
figure_dir = "/tscc/projects/ps-epigen/users/biy022/biccn/analysis/final_figures/"
g.savefig(
    "{}/calcium_genes_all_scenic_tfs_RC_RNA.pdf".format(figure_dir),
    format="pdf",
    dpi=300
)
```

```python
relevant_tfs = list(geneset_tfs & homer_positive_tfs)
plot_matrix = L4_IT_rna_region_counts[region_align, relevant_tfs].X
g = sns.clustermap(
    plot_matrix.T,
    xticklabels=region_align,
    yticklabels=relevant_tfs,
    z_score="row",
    figsize=(8, 5),
    col_cluster=False,
    dendrogram_ratio=(0.1, 0.05),
    cbar_pos=(0.5, 0.975, 0.2, 0.02),
    cbar_kws={"orientation": "horizontal"},
    tree_kws={"linewidth": 2},
    linewidth=2,
    linecolor="black",
)
ax = g.ax_heatmap
ax.tick_params(axis="x", labelsize=15)
ax.tick_params(axis="y", labelsize=15)
pos = ax.get_position()
x0 = pos.x0
y0 = pos.y0
width = pos.width
height = pos.height
rect = patches.Rectangle(
    (x0, y0),
    width,
    height,
    linewidth=2, 
    edgecolor="black", 
    facecolor="none", 
    transform=plt.gcf().transFigure
)
plt.gcf().add_artist(rect)
figure_dir = "/tscc/projects/ps-epigen/users/biy022/biccn/analysis/final_figures/"
g.savefig(
    "{}/calcium_genes_scenic_tfs_homer_positive_intersected_RC_RNA.pdf".format(figure_dir),
    format="pdf",
    dpi=300
)
```
