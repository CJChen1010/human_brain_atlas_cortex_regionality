# Calculate proportion of cell types

```python
import pandas as pd
pd.options.mode.chained_assignment = None 
import os
from pathlib import Path
```

```python
os.chdir("/tscc/projects/ps-epigen/users/biy022/biccn/data/SNAREdata")
```

```python
filtered_barcodes = list()
for file in Path("cell_peak_matrix/").glob("*.tsv"):
    curr_barcodes_table = pd.read_csv(file, sep="\t", header=None)
    curr_barcodes_table.columns = ["barcodes"]
    filtered_barcodes += curr_barcodes_table["barcodes"].to_list()
```

```python
meta_table = pd.read_csv(
    "cell_gene_matrix/20230313_RNA_metadata.xls",
    sep="\t", header=0, index_col=0
)
meta_table = meta_table.loc[filtered_barcodes, :]
```

```python
count_table = meta_table.value_counts(["Subclass_F", "Region", "PatientID"]).reset_index()
count_table.columns = ["Subclass_F", "Region", "PatientID", "count"]
```

```python
count_table["prop"] = count_table.groupby(
    ["Region", "PatientID"], group_keys=False).apply(lambda x: x["count"] / x["count"].sum())
```

```python
import seaborn as sns
```

```python
region_align = ["A24", "A9", "FI", "M1C", "S1C", "MTG", "A1C", "AnG", "V1C"]
```

```python
curr_table = count_table[count_table["Subclass_F"] == "L4 IT"]
curr_table["Region"] = pd.Categorical(curr_table["Region"], categories=region_align)
sns.stripplot(
    data=curr_table, x="Region", y="prop", hue="Region",
    dodge=False, jitter=False, alpha=.5, legend=False,
)
sns.pointplot(
    data=curr_table, x="Region", y="prop", hue="Region",
    dodge=0, linestyle="none", errorbar=None,
    marker="_", markersize=20, markeredgewidth=3,
)
```
