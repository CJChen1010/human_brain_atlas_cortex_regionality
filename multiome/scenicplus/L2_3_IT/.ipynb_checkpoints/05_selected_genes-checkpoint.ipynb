```python
import os
from pathlib import Path
root_dir = Path("/tscc/projects/ps-epigen/users/biy022/biccn/data/SNAREdata/scenicplus/")
os.chdir(root_dir / "L2_3_IT")
```

```python
import pandas as pd
import numpy as np
import math
from matplotlib_venn import venn2, venn2_circles
from matplotlib import rcParams
import matplotlib.pyplot as plt
```

```python
direct_table = pd.read_csv("scplus_outputs/eRegulon_direct.tsv", sep="\t", header=0)
extend_table = pd.read_csv("scplus_outputs/eRegulons_extended.tsv", sep="\t", header=0)
```

```python
regulons_table = pd.concat([direct_table, extend_table], axis=0)
regulons_table.reset_index(drop=True, inplace=True)
```

```python
ST_positive_peak_table = pd.read_csv(
    "../other_celltypes/corr_genes_ST_7e-1_1e-2_expr_filtered/L2_3_IT_positive.tsv",
    sep="\t", header=None
)
ST_positive_peak_table.columns = ["Region"]
ST_negative_peak_table = pd.read_csv(
    "../other_celltypes/corr_genes_ST_7e-1_1e-2_expr_filtered/L2_3_IT_negative.tsv",
    sep="\t", header=None
)
ST_negative_peak_table.columns = ["Region"]
```

```python
ST_positive_peak_table["Region"] = ST_positive_peak_table.apply(lambda x: ":".join(x["Region"].split("-", 1)), axis=1)
ST_negative_peak_table["Region"] = ST_negative_peak_table.apply(lambda x: ":".join(x["Region"].split("-", 1)), axis=1)
```

```python
receptor_subunits_table = pd.read_csv("../../share/genes_of_interest/receptor_subunits.tsv")
receptor_subunits_table.columns = ["Gene"]
```

```python
all_edges = pd.merge(regulons_table, receptor_subunits_table, on=["Gene"], how="inner")
positive_edges = pd.merge(all_edges, ST_positive_peak_table, on="Region", how="inner")
negative_edges = pd.merge(all_edges, ST_negative_peak_table, on="Region", how="inner")
```

```python
all_edges["TF"].unique()
```

```python
all_edges.to_csv("receptor_subunits_networks/L2_3_IT_network_receptor_subunits.tsv", sep="\t", header=True, index=False)
positive_edges.to_csv("receptor_subunits_networks/L2_3_IT_network_ST_positive_receptor_subunits.tsv", sep="\t", header=True, index=False)
negative_edges.to_csv("receptor_subunits_networks/L2_3_IT_network_ST_negative_receptor_subunits.tsv", sep="\t", header=True, index=False)
```

```python
motif_database = pd.read_csv(
    "/tscc/projects/ps-epigen/users/biy022/MGH/data/metadata/homer2_motif_database.csv",
    header=0, sep=","
)
motif_database = motif_database.dropna()
homer_result_positive = pd.read_csv(
    "corr_regions_ST_5e-1_5e-2_expr_filtered/positive_homer/knownResults.txt",
    header=0, sep="\t"
)
homer_result_positive = homer_result_positive[homer_result_positive["q-value (Benjamini)"] < 0.01]
homer_positive_tfs = motif_database[motif_database["motif_names"].isin(
    homer_result_positive["Motif Name"])]["tfs"].to_list()
homer_positive_tfs = set(homer_positive_tfs)
```

```python
scenicplus_tfs = set(all_edges["TF"].unique())
```

```python
figure_dir = "/tscc/projects/ps-epigen/users/biy022/biccn/analysis/final_figures/"
```

```python
rcParams["font.family"] = "Ubuntu Mono"
venn2([homer_positive_tfs, scenicplus_tfs], ("Homer", "SCENIC+"))
venn2_circles([homer_positive_tfs, scenicplus_tfs], linestyle="dashed", linewidth=2)
plt.savefig("{}/receptor_subunits_homer_positive_venn.pdf".format(figure_dir), dpi=300)
```

```python
homer_result_negative = pd.read_csv(
    "corr_regions_ST_5e-1_5e-2_expr_filtered/negative_homer/knownResults.txt",
    header=0, sep="\t"
)
homer_result_negative = homer_result_negative[homer_result_negative["q-value (Benjamini)"] < 0.01]
homer_negative_tfs = motif_database[motif_database["motif_names"].isin(
    homer_result_negative["Motif Name"])]["tfs"].to_list()
homer_negative_tfs = set(homer_negative_tfs)
```

```python
rcParams["font.family"] = "Ubuntu Mono"
venn2([homer_negative_tfs, scenicplus_tfs], ("Homer", "SCENIC+"))
venn2_circles([homer_negative_tfs, scenicplus_tfs], linestyle="dashed", linewidth=2)
plt.savefig("{}/receptor_subunits_homer_negative_venn.pdf".format(figure_dir), dpi=300)
```

```python
pd.options.mode.chained_assignment = None 
import scanpy as sc
import decoupler as dc
import numpy as np
import scipy.stats as stats
import matplotlib.patches as patches
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
## Plot the expression gradient of the intersected TFs
L2_3_IT_rna = sc.read_h5ad("L2_3_IT_rna_matrix.h5ad")
L2_3_IT_rna_counts = L2_3_IT_rna.raw.to_adata()
L2_3_IT_rna_region_counts = dc.get_pseudobulk(
    adata=L2_3_IT_rna_counts,
    sample_col="Region",
    groups_col=None,
    min_cells=0,
    min_counts=0,
    min_prop=0.0,
    min_smpls=0,
    remove_empty=True,
    mode="sum"
)
sc.pp.normalize_total(L2_3_IT_rna_region_counts, target_sum=1e6)
sc.pp.log1p(L2_3_IT_rna_region_counts, base=2)
region_align =["A24", "A9", "FI", "MTG", "AnG", "M1C", "S1C", "A1C", "V1C"]
```

```python
relevant_tfs = list(scenicplus_tfs)
plot_matrix = L2_3_IT_rna_region_counts[region_align, relevant_tfs].X
g = sns.clustermap(
    plot_matrix.T,
    xticklabels=region_align,
    yticklabels=relevant_tfs,
    z_score="row",
    figsize=(8, 16),
    col_cluster=False,
    dendrogram_ratio=(0.1, 0.05),
    cbar_pos=(0.5, 0.975, 0.2, 0.02),
    cbar_kws={"orientation": "horizontal"},
    tree_kws={"linewidth": 2},
    linewidth=2,
    linecolor="black",
)
ax = g.ax_heatmap
ax.tick_params(axis="x", labelsize=15)
ax.tick_params(axis="y", labelsize=15)
pos = ax.get_position()
x0 = pos.x0
y0 = pos.y0
width = pos.width
height = pos.height
rect = patches.Rectangle(
    (x0, y0),
    width,
    height,
    linewidth=2, 
    edgecolor="black", 
    facecolor="none", 
    transform=plt.gcf().transFigure
)
plt.gcf().add_artist(rect)
figure_dir = "/tscc/projects/ps-epigen/users/biy022/biccn/analysis/final_figures/"
g.savefig(
    "{}/receptor_subunits_all_scenic_tfs_RC_RNA.pdf".format(figure_dir),
    format="pdf",
    dpi=300
)
```

```python
relevant_tfs = ["KLF12", "RFX3", "RFX2"]
plot_matrix = L2_3_IT_rna_region_counts[region_align, relevant_tfs].X
g = sns.clustermap(
    plot_matrix.T,
    xticklabels=region_align,
    yticklabels=relevant_tfs,
    z_score="row",
    figsize=(8, 1.25),
    col_cluster=False,
    dendrogram_ratio=(0.1, 0.05),
    cbar_pos=(0.5, 0.975, 0.2, 0.02),
    cbar_kws={"orientation": "horizontal"},
    tree_kws={"linewidth": 2},
    linewidth=2,
    linecolor="black",
)
ax = g.ax_heatmap
ax.tick_params(axis="x", labelsize=15)
ax.tick_params(axis="y", labelsize=15)
pos = ax.get_position()
x0 = pos.x0
y0 = pos.y0
width = pos.width
height = pos.height
rect = patches.Rectangle(
    (x0, y0),
    width,
    height,
    linewidth=2, 
    edgecolor="black", 
    facecolor="none", 
    transform=plt.gcf().transFigure
)
plt.gcf().add_artist(rect)
figure_dir = "/tscc/projects/ps-epigen/users/biy022/biccn/analysis/final_figures/"
g.savefig(
    "{}/receptor_subunits_selected_scenic_tfs_RC_RNA.pdf".format(figure_dir),
    format="pdf",
    dpi=300
)
```
